```python
# Part 1: Predicting next word
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from collections import defaultdict, Counter
```

```python
#Initialize
text = open("dataset.txt", "r").read()
lookup_dict = defaultdict(list)
cleaned = re.sub(r'\W+', ' ', text).lower()
tokenized = word_tokenize(cleaned)
```

```python
def generate_tuple_keys(data):
    if len(data) < 1:
        return
    for i in range(len(data) - 1):
        yield [ data[i], data[i + 1] ]
```

```python
pairs = generate_tuple_keys(tokenized)
for pair in pairs:
    lookup_dict[pair[0]].append(pair[1])
```

```python
# test
input_word = input().lower()
try:
    print(input_word, Counter(lookup_dict[input_word]).most_common()[0])
    print(input_word, Counter(lookup_dict[input_word]).most_common()[1])
    print(input_word, Counter(lookup_dict[input_word]).most_common()[2])
except IndexError:
    print('No suggestion')
```

```python
import gzip
import gensim 
import logging

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
def read_input(input_file):
    """This method reads the input file which is in gzip format"""
    
    logging.info("reading file {0}...this may take a while".format(input_file))
    
    with gzip.open (input_file, 'rb') as f:
        for i, line in enumerate (f): 

            if (i%10000==0):
                logging.info ("read {0} reviews".format (i))
            # do some pre-processing and return a list of words for each review text
            yield gensim.utils.simple_preprocess (line)

# read the tokenized reviews into a list
# each review item becomes a serries of words
# so this becomes a list of lists

data_file="reviews_data.txt.gz"

documents = list (read_input (data_file))
logging.info ("Done reading data file")
```

```python
model = gensim.models.Word2Vec (documents, size=150, window=10, min_count=2, workers=10)
model.train(documents,total_examples=len(documents),epochs=10)
```

```python
w1 = "small"
model.wv.most_similar (positive=w1)
```

```python
input_word = input().lower()
try:
    most_common = Counter(lookup_dict[input_word]).most_common()[0][0]
    print(most_common)
    print(model.wv.most_similar (positive=most_common))
except IndexError:
    print('No suggestion')
```
